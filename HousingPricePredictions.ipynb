```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()
data = pd.read_csv('housing.csv')
```

```python
data.info() #check if an null values, we null in total bedrooms-->as less values null-drop, data type has one object,
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
#encoding
data = pd.get_dummies(data, columns=['ocean_proximity'], drop_first=True)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split

x = data.drop(["median_house_value"], axis=1)
y = data["median_house_value"]
```

```python
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=0.2,)
```

```python
#find correlation between cols of training set
train_data = x_train.join(y_train)
train_data.hist(figsize=(15,10))
```

```python
train_data
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(train_data.corr(), annot = True, cmap="YlGnBu")
```

```python
#for the skewed/asymmetrical data from hist, log used as right-skewed, in order to not get a biased model
train_data["total_rooms"] = np.log(train_data["total_rooms"]+1) #to avoid 0 values,+1
train_data["total_bedrooms"] = np.log(train_data["total_bedrooms"]+1)
train_data["population"] = np.log(train_data["population"]+1)
train_data["households"] = np.log(train_data["households"]+1)
```

```python
train_data.hist(figsize=(15,10))
```

```python
#to see how position of individual blocks helps with pricing --> scatterplot
plt.figure(figsize=(15,8))
sns.scatterplot(x="longitude", y="latitude", data=train_data, hue="median_house_value", palette="coolwarm")
```

```python
#feature engineering to get interesting features
#price acc to the total bedrooms they offer; price acc to how many rooms a household/family has

train_data["rooms_per_household"] = train_data["total_rooms"]/train_data["households"]
train_data["bedrooms_per_room"] = train_data["total_bedrooms"]/train_data["total_rooms"]

plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot = True, cmap="YlGnBu")
```

LINEAR REGRESSION


```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

x_train = train_data.drop(["median_house_value"], axis=1)
y_train = train_data["median_house_value"]

x_train_s = StandardScaler().fit_transform(x_train) #usually do not need to scale outputs

model = LinearRegression()
model.fit(x_train_s, y_train)
```

```python
#testing

test_data = x_test.join(y_test)

test_data["total_rooms"] = np.log(test_data["total_rooms"]+1) #to avoid 0 values,+1
test_data["total_bedrooms"] = np.log(test_data["total_bedrooms"]+1)
test_data["population"] = np.log(test_data["population"]+1)
test_data["households"] = np.log(test_data["households"]+1)

test_data["rooms_per_household"] = test_data["total_rooms"]/test_data["households"]
test_data["bedrooms_per_room"] = test_data["total_bedrooms"]/test_data["total_rooms"]
```

```python
x_test = test_data.drop(["median_house_value"], axis=1)
y_test = test_data["median_house_value"]

x_test_s = StandardScaler().fit_transform(x_test)
model.score(x_test_s, y_test)
```

RANDOM FOREST

```python
#special type of ml where we combine diff model
```

```python
from sklearn.ensemble import RandomForestRegressor

model2 = RandomForestRegressor()
model2.fit(x_train_s, y_train)
```

```python
model2.score(x_test_s, y_test)
```

```python
#try to get better accuracy, may or may not, by providing a parameter grid
#use grid search for cross validation

from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()

param_grid = {
    "n_estimators": [10, 40, 90],
    "max_depth":[None, 6, 15]
}

grid_search_model = GridSearchCV( forest, param_grid, cv = 5,
                                 scoring= "neg_mean_squared_error",
                                  return_train_score = True)

grid_search_model.fit(x_train_s, y_train)
```

```python
grid_search_model.best_estimator_
```

```python
grid_search_model.best_estimator_.score(x_test_s, y_test)
```
